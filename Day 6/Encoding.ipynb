# Label Encoder using Loop

```python
import pandas as pd
df = pd.read_csv('insurance.csv')
df.head()
```

```python
import numpy as np

import warnings
warnings.filterwarnings('ignore')

for column in df.columns:
    if df[column].dtype==np.number:
        continue
    df[column] = le.fit_transform(df[column])
```

# One Hot Encoding

```python
import pandas as pd
df = pd.read_csv('insurance.csv')
df.head()
```

```python
dummy = pd.get_dummies(df['region'])
```

```python
dummy.head()
```

```python
columns = ['region']

for col in columns:
    one_hot = pd.get_dummies(df[col])
    df = pd.concat((df, one_hot), axis=1)
```

```python
df.head()
```

# Ordinal Using Loop

```python
import pandas as pd
df = pd.read_csv('insurance.csv')
df.head()
```

```python
df.region.unique()
```

```python
country = ['southwest', 'southeast', 'northwest', 'northeast']
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
ordinal = OrdinalEncoder(categories=[country])
```

```python
encoded = ordinal.fit_transform(df[['region']])
```

```python
newdata_frame = pd.DataFrame(encoded, columns=['newRegion'])
```

```python
newdata_frame
```

```python
df.region = newdata_frame.newRegion
```

```python
df.head()
```

```python
df.sex.unique()
```

```python
sex = ['female', 'male']
```

```python
ordinal = OrdinalEncoder(categories=[sex])
```

```python
encoded1 = ordinal.fit_transform(df[['sex']])
```

```python
new_dataframe = pd.DataFrame(encoded1, columns=['NewSex'])
```

```python
new_dataframe.head()
```

```python
df.head()
```

```python
df.sex = new_dataframe.NewSex
```
